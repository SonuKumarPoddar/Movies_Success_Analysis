```python
import ast

country_map = {
    'United States of America': 'USA',
    'United Kingdom': 'UK',
    'India': 'India',
    'France': 'France',
    'Germany': 'Germany',
    'Canada': 'Canada',
    'Japan': 'Japan',
    'China': 'China'
}

def clean_country(country_text):
    try:
        countries = ast.literal_eval(country_text)
        names = [country_map.get(c['name'], c['name']) for c in countries]
        return ', '.join(names[:2])   # keep max 2
    except:
        return None
```

```python
df['country_clean'] = df['production_countries'].apply(clean_country)
```

```python
def clean_company(company_text):
    try:
        companies = ast.literal_eval(company_text)
        names = []
        for c in companies[:2]:
            short_name = ' '.join(c['name'].split()[:2])
            names.append(short_name)
        return ', '.join(names)
    except:
        return None
```

```python
df['production_company_clean'] = df['production_companies'].apply(clean_company)
```

```python
def popularity_category(value):
    if value < 10:
        return 'Very Low'
    elif value < 30:
        return 'Low'
    elif value < 60:
        return 'Medium'
    elif value < 100:
        return 'High'
    else:
        return 'Very High'
```
